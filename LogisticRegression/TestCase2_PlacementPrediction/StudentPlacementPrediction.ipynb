## 3. Data Cleaning

```python
student_data.describe()
```

```python
student_gender = pd.get_dummies(student_data["gender"],drop_first=True, prefix="gender" )
work_experience = pd.get_dummies(student_data["workex"],drop_first=True, prefix="WorkEx")
placement_status = pd.get_dummies(student_data["status"].values,drop_first=True, prefix="Placement" )
```

```python
student_data = pd.get_dummies(data=student_data, columns=["ssc_b","hsc_b", "hsc_s","degree_t","specialisation"])
student_data = pd.concat([student_data,student_gender,work_experience,placement_status], axis =1)
student_data
```

```python
student_data.drop(["sl_no","gender","workex","status","salary"],axis=1,inplace = True)
```

```python
student_data.isnull().sum()
```

```python
student_data.info()
```

## 4. Split Dataset

```python
X = student_data.drop(["Placement_Placed"],axis=1)
Y = student_data["Placement_Placed"]
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

## 5. Modeling

```python
from sklearn.linear_model import LogisticRegression
logReg = LogisticRegression(max_iter=1000)
logReg.fit(X_train,Y_train)
```

## 6. Testing

```python
Y_predict = logReg.predict(X_test)
Y_predict
```

## 7. Accuracy Checking

```python
from sklearn.metrics import accuracy_score
print("Accuracy:",accuracy_score(Y_test,Y_predict)*100)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_predict))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test,Y_predict))
```
